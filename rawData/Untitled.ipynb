```python
pers = tt._shared(np.arange(48*4).reshape(48,4))
pers.eval()
```

```python
choice = tt._shared(np.random.randint(low=0, high=4, size=48))
tt.shape(choice).eval()
choice.eval()
```

```python
pers[tt.arange(choice.shape[0]), choice].eval()
```

```python
pos = tt._shared(np.ones(48))
neg = np.full(48, -1)
neg = tt._shared(neg)
```

```python
outcome = tt._shared(np.random.randint(low=-1, high=1, size=(48,10)))
outcome.eval()
```

```python
update_pers = tt.switch(outcome[:, 1] >= 0, pos, neg)
update_pers.eval()
```

```python
pers = tt.inc_subtensor(pers[tt.arange(choice.shape[0]), choice], update_pers)
```

```python
pers.eval()
```

```python
tt.shape(outcome[:, 1]).eval()
```

```python
phi = tt._shared(np.arange(48).reshape(48,))
ev = tt._shared(np.arange(48*4).reshape(48,4))
```

```python
util = tt._shared(np.arange(48))
```

```python
inc = phi * (util - ev[tt.arange(choice.shape[0]), choice])
inc.eval()
```

```python
ev = tt.inc_subtensor(ev[tt.arange(choice.shape[0]), choice], inc)
```

```python
ev.eval()
```
